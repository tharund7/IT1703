```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

# Load the dataset


```python
# Loading the data
data = pd.read_csv('breast-cancer.csv')

# Checking Dataset
data
```

# Define feature columns and target column


```python
features = data.columns[1:-1]
target = 'Class'
```

# Split the data into training and testing sets


```python
X_train, X_test, y_train, y_test = train_test_split(data[features], data[target], test_size=0.2, random_state=42)
```

# Function to perform hyperparameter tuning and evaluation


```python
def train_evaluate_svm(model, param_grid):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_
```

# Cross-validation training accuracy


```python
def train_evaluate_svm(model, param_grid):
    grid_search = GridSearchCV(model, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    best_model = grid_search.best_estimator_

    # Cross-validation training accuracy
    cv_scores = cross_val_score(best_model, X_train, y_train, cv=5)
    cv_accuracy = cv_scores.mean()

    # Testing accuracy
    test_predictions = best_model.predict(X_test)
    test_accuracy = accuracy_score(y_test, test_predictions)

    return best_model, cv_accuracy, test_accuracy
```

# Linear SVM


```python
linear_svm = SVC(kernel='linear')
linear_param_grid = {'C': [0.1, 1, 10, 100]}
linear_best_model, linear_cv_accuracy, linear_test_accuracy = train_evaluate_svm(linear_svm, linear_param_grid)
```

# RBF Kernel SVM


```python
rbf_svm = SVC(kernel='rbf')
rbf_param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [0.001, 0.01, 0.1, 1]}
rbf_best_model, rbf_cv_accuracy, rbf_test_accuracy = train_evaluate_svm(rbf_svm, rbf_param_grid)
```

# Linear Kernel SVM


```python
linear_kernel_svm = SVC(kernel='linear')
linear_kernel_best_model, linear_kernel_cv_accuracy, linear_kernel_test_accuracy = train_evaluate_svm(linear_kernel_svm, linear_param_grid)
```

# Report the results


```python
results = {
    "Model": ["Linear SVM", "RBF Kernel SVM", "Linear Kernel SVM"],
    "CV Accuracy": [linear_cv_accuracy, rbf_cv_accuracy, linear_kernel_cv_accuracy],
    "Test Accuracy": [linear_test_accuracy, rbf_test_accuracy, linear_kernel_test_accuracy]
}
results_df = pd.DataFrame(results)
print(results_df)
```
